```python
import pandas as pd


# Delete the columns Time, Sub_metering_2 and Sub_metering_3

df = pd.read_csv("./data/household_power_consumption.txt", sep=";")
df.drop(["Time", "Sub_metering_2", "Sub_metering_3"], inplace=True, axis=1)
df
```

```python
# Set Date as index

df.set_index("Date", inplace=True)
df.index = pd.to_datetime(df.index, dayfirst=True)
df.index = df.index.astype("datetime64[ns]")
# df.index = df.index.strftime("%Y-%m-%d %H:%M:%S")
print(df.index.dtype)
df
```

```python
# DataFrame with updated types:


def update_types(df):
    df = df.replace("?", "nan")
    df["Global_active_power"] = df["Global_active_power"].astype(float)
    df["Global_reactive_power"] = df["Global_reactive_power"].astype(float)
    df["Voltage"] = df["Voltage"].astype(float)
    df["Global_intensity"] = df["Global_intensity"].astype(float)
    df["Sub_metering_1"] = df["Sub_metering_1"].astype(float)
    return df

df = update_types(df)
print(df.dtypes)
```

```python
df.describe()
```

```python
# Delete the rows with missing values

print("before:\n",df.isna().sum())

df.dropna(inplace=True)

print("----\nafter\n",df.isna().sum())
# df
```

```python
# Modify Sub_metering_1 by adding 1 to it and multiplying the total by 0.06.
# (x+1)*0.06

df.loc[:, "Sub_metering_1"] = (df["Sub_metering_1"] + 1) * 0.06

print(df["Sub_metering_1"].head())
```

```python
# Select all the rows for which the Date >= 2008-12-27, and Voltage >= 242
res = df[(df.index >= "2008-12-27") & (df["Voltage"] >= 242.0)]
print(res.shape)
res
print(res[["Global_active_power", "Global_reactive_power"]].head().to_markdown())
```

```python
# Sort the first three columns by descending order of Global_active_power and ascending order of Voltage

three_col = df.iloc[:, :3]
sorted = three_col.sort_values(by=["Global_active_power", "Voltage"], ascending=[False, True])
print(sorted.tail().to_markdown())
```

```python
# Compute the daily average of Global_active_power.

d_avg = df["Global_active_power"].groupby(df.index).mean()
print(d_avg)
```
