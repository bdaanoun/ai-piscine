```python
# Regression example
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.layers import Dense

data = pd.read_csv("./data/auto-mpg.csv")
df = pd.DataFrame(data)
df= df.drop(["car name", "model year", "origin"], axis=1)

df.replace("?", pd.NA, inplace=True)
df.dropna(inplace=True)


# defing x and y
X = df.drop(["mpg"], axis=1)
y = df["mpg"]
print("the target\n",y[:5])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

scaler = StandardScaler()

scalled_train = scaler.fit_transform(X_train)
print("x train shape: ",scalled_train.shape)
print(scalled_train[:5])
scalled_test = scaler.transform(X_test)
print("x test shape: ", scalled_test.shape)
print("", scalled_test[:5])

```

```python
from keras.models import Sequential

model = Sequential()

model.add(Dense(64, input_dim=5, activation="sigmoid"))
model.add(Dense(32, activation="sigmoid"))

#output neuron
model.add(Dense(1))
model.compile(optimizer="adam", loss="mean_squared_error", metrics=["mean_absolute_error"])

model.fit(scalled_train, y_train, epochs=200)
y_pred = model.predict(scalled_test)
loss, mae = model.evaluate(scalled_test, y_test)
print("loss :", loss)
print("mae :", mae)
```
