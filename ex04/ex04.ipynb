```python
import pandas as pd

# Preliminary: Drop the flower column
df = pd.read_csv("./data/iris.csv")
col = df.columns
print(col)
df.drop("flower", axis=1)
```

```python
# Fill the missing values with : sepal_length -> mean
# Fill the missing values with : sepal_width -> median
# Fill the missing values with : petal_length, petal_width -> 0


# pd.to_numeric(df.loc[:,col], errors='coerce')
df["sepal_length"] = pd.to_numeric(df["sepal_length"], errors="coerce")
df["sepal_width"] = pd.to_numeric(df["sepal_width"], errors="coerce")
df["petal_length"] = pd.to_numeric(df["petal_length"], errors="coerce")
df["petal_width"] = pd.to_numeric(df["petal_width"], errors="coerce")

# df["sepal_length"].fillna(df["sepal_length"].mean())
# df["sepal_width"].fillna(df["sepal_width"].median())
# df["petal_length"].fillna(0)
# df["petal_length"].fillna(0)
# df

df.fillna({
    "sepal_length" : df["sepal_length"].mean(),
    "sepal_width" : df["sepal_width"].median(),
    "petal_length" : 0,
    "petal_width" : 0,
})
df.describe()
```
