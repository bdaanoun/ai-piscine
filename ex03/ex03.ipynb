```python
import pandas as pd



# how many rows and cols
data = pd.read_csv("./data/Ecommerce_purchases.txt", sep=",")

print(data.columns)
rows, cols = data.shape
print("rows= ", rows, "\ncols= ", cols)
```

```python
# the average Purchase Price?


avg = data["Purchase Price"].mean()
print(avg)
```

```python
# the highest and lowest purchase prices?

min = data["Purchase Price"].min()
max = data["Purchase Price"].max()
print("min= ", min, "\nmax= ", max)
```

```python
# 'en' as their Language

en_lang = (data["Language"] == "en").sum()
print(en_lang)
```

```python
#  the job title of "Lawyer"

job_title_count =(data["Job"] == "Lawyer").sum()
print(job_title_count)
```
