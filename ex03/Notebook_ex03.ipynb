```python
import spacy
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics.pairwise import cosine_similarity

nlp = spacy.load("en_core_web_md")

text = "laptop computer coffee tea water liquid dog cat kitty"
doc = nlp(text)

words = [token.text for token in doc]
vectors = np.array([token.vector for token in doc])

print("Shape of laptop vector:", vectors[0].shape)

print("Number of words:", len(words))
print("Words:", words)


laptop_sum = sum(vectors[0][:20])
computer_sum = sum(vectors[1][:20])

print("Laptop sum:", laptop_sum)

similarity_matrix = 1- cosine_similarity(vectors)

plt.figure(figsize=(10, 6))

plt.imshow(
    similarity_matrix[::-1],  # Reverse the order of rows for correct y-axis labels
    cmap="plasma",
    aspect="auto"
)

plt.xticks(range(len(words)), words)
plt.yticks(range(len(words)), words[::-1])

plt.title("Cosine distance between words")

plt.colorbar()

plt.tight_layout()

plt.savefig("w3day05ex1_plot.png")

# Show
plt.show()
```
