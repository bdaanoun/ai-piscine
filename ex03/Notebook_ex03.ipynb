```python
# OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

X_train = [['Python'], ['Java'], ['Java'], ['C++']]

encoder = OneHotEncoder(handle_unknown='ignore')
encoder.fit(X_train)
encoder_X = encoder.transform(X_train)

cols = encoder.categories_

df = pd.DataFrame(encoder_X.toarray(), columns=cols)
print(df.to_markdown())

```

```python
# Transform X_test using the OneHotEncoder

X_test = [['Python'], ['Java'], ['C'], ['C++']]

encoder_X_test = encoder.transform(X_test)

cols = encoder.categories_

df = pd.DataFrame(encoder_X_test.toarray(), columns=cols)

print(df.to_markdown())
```
