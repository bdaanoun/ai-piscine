```python
import numpy as np
from sklearn.metrics import log_loss
y_true = np.array([0,1,1,0,1])
y_pred = np.array([0.1,0.8,0.6, 0.5, 0.3])

# Log loss: - 1/n * Sum[(y_true*log(y_pred) + (1-y_true)\*log(1-y_pred))]

log_loss_val = -1/len(y_true) * np.sum((y_true)*np.log(y_pred)+ (1 - y_true) * np.log(1-y_pred))
print("log loss using formula",log_loss_val)
log_l = log_loss(y_true, y_pred)
print("log loss using sklearn",log_l)
```
