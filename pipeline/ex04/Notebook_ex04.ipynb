```python
# Ordinal Encoder
from sklearn.preprocessing import OrdinalEncoder

X_train = [['good'], ['bad'], ['neutral']]

catgs=[['bad', 'neutral', 'good']]

enc = OrdinalEncoder(categories=catgs)
enc.fit(X_train)

transX = enc.fit_transform(X_train)

print(transX)

print(enc.categories_)
```

```python
X_test = [['good'], ['good'], ['bad']]

transX_test = enc.transform(X_test)
transX_test
```
