```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.compose import make_column_transformer

df = pd.read_csv('data/breast-cancer.csv',header=None)
df.columns = [
    "age",
    "menopause",
    "tumor-size",
    "inv-nodes",
    "node-caps",
    "deg-malig",
    "breast",
    "breast-quad",
    "irradiat",
    "Class"
]
# Drop Class column and NaN values
df = df.drop(columns=['Class'])
df = df.dropna()

X_train, X_test = train_test_split(df, test_size=0.2, random_state=43)

print(X_train.nunique())
```

```python

ohe_cols = ['node-caps', 'breast', 'breast-quad', 'irradiat']

ohe = OneHotEncoder(sparse_output=False)   # sklearn >= 1.2 uses sparse_output
ohe.fit(X_train[ohe_cols])
print(ohe.transform(X_test[ohe_cols])[:10])

print("\nFeature names:")
print(ohe.get_feature_names_out(ohe_cols))
```

```python


ord_cols = ["menopause", "age", "tumor-size", "inv-nodes", "deg-malig"]

categories = [
    ['lt40', 'premeno', 'ge40'],    ['10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99'],
    ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34',
     '35-39', '40-44', '45-49', '50-54', '55-59'],
    ['0-2', '3-5', '6-8', '9-11', '12-14', '15-17', '18-20',
     '21-23', '24-26', '27-29', '30-32', '33-35', '36-39'],
    [1, 2, 3],
]

oe = OrdinalEncoder(categories=categories)
oe.fit(X_train[ord_cols])

print(oe.transform(X_test[ord_cols])[:10])
```

```python

ct = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ohe_cols),
    (OrdinalEncoder(categories=categories), ord_cols),
)
ct.fit(X_train)
print(ct.transform(X_test)[:2])
```
