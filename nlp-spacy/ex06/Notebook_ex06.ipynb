```python
# Part-of-speech tags
import spacy

nlp = spacy.load("en_core_web_sm")
with open("data/news_amazon.txt") as f:
    text = f.read()


doc = nlp(text)
for sent in doc.sents:
    for token in sent:
        if token.text == "Bezos" and token.pos_ == "PROPN":
            print("INFO:",token.text, token.pos_, token.tag_)
            print("Sentence:", sent.text)
            print()
```
