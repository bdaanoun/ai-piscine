```python
import pandas as pd


#df1

df1_dict = {
        'id': ['1', '2', '3', '4', '5'],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']}

df1 = pd.DataFrame(df1_dict, columns = ['id', 'Feature1', 'Feature2'])

#df2
df2_dict = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}

df2 = pd.DataFrame(df2_dict, columns = ['id', 'Feature1', 'Feature2'])

# Merge

# cd = pd.concat((df1.head(2), df2["Feature1"].head(2), df2["Feature2"].head(2)), axis=1)
# cd.columns = [ "id" , "Feature1_x"   , "Feature2_x"   , "Feature1_y"   , "Feature2_y"  ]

cd = pd.merge(df1, df2, on="id", how="inner")
print(cd.to_markdown())
```

```python
# outer join 

merged = pd.merge(df1, df2, on="id", how="outer", suffixes=("_df1","_df2") )

print(merged.to_markdown())
```
