```python
import pandas as pd

# 1. Create the Series
date_range = pd.date_range(start='2010-01-01', end='2020-12-31', freq='D')
integer_series = pd.Series(range(len(date_range)), date_range)

print("integer_series")
print(integer_series.head(10))
print(f"...\nLength: {len(integer_series)}")

# 2. Compute 7-day moving average
moving_avg = integer_series.rolling(window=7).mean()

print()
print("7-Day Moving Average" )
print(moving_avg.head(10))
```
