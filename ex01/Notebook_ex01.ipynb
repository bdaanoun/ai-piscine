```python
import numpy as np
from sklearn.impute import SimpleImputer

train_data = [[7, 6, 5],
              [4, np.nan, 5],
              [1, 20, 8]]


impl_mean = SimpleImputer(strategy="mean")

impl_mean.fit(train_data)
print("statistics_",impl_mean.statistics_)

print("nanmean",np.nanmean(train_data, axis=0))
# print((train_data))
```

```python
# Fill the missing values

filled_train_data = impl_mean.transform(train_data)
filled_train_data
```

```python
# Fill the missing values in test_data

test_data = [[np.nan, 1, 2],
             [7, np.nan, 9],
             [np.nan, 2, 4]]

filled_test_data = impl_mean.transform(test_data)
filled_test_data
```
