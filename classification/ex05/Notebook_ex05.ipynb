```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

column_names = [
    'Sample code number', 'Clump Thickness', 'Uniformity of Cell Size',
    'Uniformity of Cell Shape', 'Marginal Adhesion',
    'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin',
    'Normal Nucleoli', 'Mitoses', 'Class'
]

df = pd.read_csv(
    './data/breast-cancer-wisconsin.data',
    header=None,
    names=column_names,
    na_values='?'
)

df['Bare Nuclei'] = df['Bare Nuclei'].astype(float)
df.fillna(df.median(), inplace=True)

df.drop(columns=['Sample code number'], inplace=True)

benign_proportion = (df['Class'] == 2).mean()
print(f"Proportion of Benign: ", benign_proportion)
print(f"Accuracy if always predicts Benign: ", benign_proportion)

X = df.drop(columns=['Class'])
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=43
)

model = LogisticRegression(solver='liblinear', max_iter=1000)
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print(f"Train accuracy: ", train_score)
print(f"Test accuracy: ", test_score)

cm_train = confusion_matrix(y_train, y_train_pred, labels=[2, 4])
cm_test = confusion_matrix(y_test, y_test_pred, labels=[2, 4])

print("Confusion matrix (train):")
print(cm_train)
print("Confusion matrix (test):")
print(cm_test)

tn_train, fp_train, fn_train, tp_train = cm_train.ravel()
tn_test, fp_test, fn_test, tp_test = cm_test.ravel()

print(f"Train false negatives: {fn_train}, false positives: {fp_train}")
print(f"Test false negatives: {fn_test}, false positives: {fp_test}")
```
