```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid1(x):
    return 1 / (1 + np.exp(-(0.5 * x + 3)))

def sigmoid2(x):
    return 1 / (1 + np.exp(-(5 * x + 11)))


x  = np.linspace(-20, 15, 300)
y0 = sigmoid(x)
y1 = sigmoid1(x)
y2 = sigmoid2(x)

plt.figure(figsize=(9, 5))

plt.plot(x, y1, label='sigmoid1', color='orange', linewidth=2)
plt.plot(x, y2, label='sigmoid2', color='green',  linewidth=2)
plt.plot(x, y0, label='sigmoid',  color='blue',   linewidth=2)
plt.axhline(0.5, color='red', linestyle='-', linewidth=1.5, label='threshold = 0.5')

plt.xlabel('x')
plt.title('Sigmoids')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
