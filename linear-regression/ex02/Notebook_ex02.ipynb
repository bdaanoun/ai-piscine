```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# Linear regression in 1D

X, y, coef = make_regression(n_samples=100,
                         n_features=1,
                         n_informative=1,
                         noise=10,
                         coef=True,
                         random_state=0,
                         bias=100.0)

print(coef)
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("X (1 dimension) and y")

plt.show()
```

```python
# y = coef * x + intercept

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, y)
print(f"y = {model.coef_} * x + {model.intercept_}")

y_pred = model.predict(X)

plt.plot(X, y_pred, color="red")
plt.scatter(X, y)

plt.xlabel("X")
plt.ylabel("y")
plt.title("X (1 dimension) and y")

plt.show()
```

```python
# Predict on X.

print(y_pred[:10])
```

```python
# computes the Mean Squared Error (MSE)
def compute_mse(y_true, y_pred):
    nsample = len(y_true)
    error = y_true - y_pred
    squared_err = error**2

    mse = squared_err.sum() / nsample
    return mse
print("MSE=", compute_mse(y, y_pred))
```
