```python
import numpy as np

from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Load dataset
data = fetch_california_housing()

X = data.data
y = data.target

# Model
model = RandomForestRegressor(random_state=42)

# Parameter grid
param_grid = {
    "n_estimators": [50, 100, 150],
    "max_depth": [10, 20, None],
    "min_samples_leaf": [1, 2, 4]
}

# Custom CV split
cv_split = [(np.arange(18576), np.arange(18576, 20640))]

# Grid Search
gridsearch = GridSearchCV(
    model,
    param_grid,
    cv=[(np.arange(18576), np.arange(18576,20640))],
    n_jobs=-1
)


gridsearch.fit(X, y)

# Function required
def select_model_verbose(gs):

    return gs.best_estimator_, gs.best_params_, gs.best_score_


trained_model, best_params, best_score = select_model_verbose(gridsearch)

# Predict new point
new_point = np.array([[3.2031, 52., 5.47761194, 1.07960199, 910., 2.26368159, 37.85, -122.26]])


prediction = trained_model.predict(new_point)

print("Prediction:", prediction)
```
