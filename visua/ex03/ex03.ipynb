```python
import matplotlib.pyplot as plt
import numpy as np

x= [1,4,5,6,7]
y= [2,6,3,6,3]
print(x, y)

page, ax = plt.subplots()

# red dasheched ...
ax.plot(x,y, linestyle="-.", color="red", linewidth=3)


# blue circles 
ax.plot(x,y, marker="o", color="blue", markersize=12, linestyle="None")
ax.set_title("Display marker")
ax.set_xlabel("x - axis")
ax.set_ylabel("y - axis")
plt.xlim(1,8)
plt.ylim(1,8)
plt.show()
```
