```python
# Regression - Optimize
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(8, input_shape=(5,), activation= 'sigmoid'))
model.add(Dense(4, activation= 'sigmoid'))
model.add(Dense(1, activation= 'linear'))

model.compile(
  optimizer='adam',
  loss='mse',
  metrics=['mse']
)


```
