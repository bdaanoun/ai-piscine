```python
import math

class Neuron:
    def __init__(self, weight1, weight2, bias, regression):
        self.weight1 = weight1
        self.weight2 = weight2
        self.bias = bias
        self.regression = regression

    def sigmoid(self, x):
        return 1 / (1 + math.exp(-x))

    def feedforward(self, x1, x2):

        total = (
            x1 * self.weight1 +
            x2 * self.weight2 +
            self.bias
        )

        if self.regression:
            return total
        else:
            return self.sigmoid(total)



neuron = Neuron(0, 1, 4, True)

# print(neuron.feedforward(2, 3))

regression = True

neuron_h1 = Neuron(0.05, 0.001, 0, False)
neuron_h2 = Neuron(0.002, 0.003, 0, False)
neuron_o1 = Neuron(2, 7, 10, True)


def predict(x1, x2):

    h1 = neuron_h1.feedforward(x1, x2)
    h2 = neuron_h2.feedforward(x1, x2)

    o1 = neuron_o1.feedforward(h1, h2)

    return o1


students = [
    ("Bob", 12, 15, 16),
    ("Eli", 10, 9, 10),
    ("Tom", 18, 18, 19),
    ("Ryan", 13, 14, 16)
]

predictions = []
targets = []

for name, math_g, chem_g, physics_g in students:

    pred = predict(math_g, chem_g)

    print(name, ":", pred)

    predictions.append(pred)
    targets.append(physics_g)


mse = 0

for y_true, y_pred in zip(targets, predictions):
    mse += (y_true - y_pred) ** 2

mse = mse / len(targets)

print("MSE:", mse)
```
