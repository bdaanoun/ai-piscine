```python
# The neuron

import numpy as np

class Neuron:
    def __init__(self, weight1, weight2, bias):
        self.weights_1 = weight1
        self.weights_2 = weight2
        self.bias = bias

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def feedforward(self, x1, x2):
        weighted_sum = (x1 * self.weights_1) + (x2 * self.weights_2) + self.bias
        
        y = self.sigmoid(weighted_sum)
        return y


neuron = Neuron(weight1=0, weight2=1, bias=4)
output = neuron.feedforward(x1=2, x2=3)
print("Output:", output)
```
