```python
import numpy as np
import pandas as pd
import plotly.express as px

returns = np.random.randn(50)
price = 100 + np.cumsum(returns)
dates = pd.date_range(start='2020-09-01', periods=50, freq='B')
df = pd.DataFrame(zip(dates, price), columns=['Date', 'Company_A'])
df

fig = px.line(df, 
              x='Date', 
              y='Company_A',
              title='Company_1 stock price')

fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Price'
)

fig.show()
```

```python
# using plotly.graph_objects

import plotly.graph_objects as go

# print(df)
fig = go.Figure()
fig.add_trace(go.Scatter(x=df["Date"], y=df["Company_A"], mode="lines"))

fig.update_layout(title="Company_1 stock price", xaxis_title="Date", yaxis_title="Price")

fig.show()
```
